```python
def plot_random(data_array,cols=5,rows=5):
    num_images = data_array.shape[0]
    fig,axes = plt.subplots(rows,cols,figsize=(20,20))
    random_choices = np.arange(0,100,1)

    for i in range(rows):
        for j in range(cols):
            axes[i][j].imshow(data_array[random_choices[(i+1)*(j+1)-1]].reshape(250,100))

    plt.show()
```

```python
plot_random(data_transpose)
```
